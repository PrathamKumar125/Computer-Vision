```python
cr=img[20:100,10:100]
plt.subplot(1,2,1);plt.imshow(cr,cmap='gray')
plt.subplot(1,2,2);plt.imshow(img,cmap='gray')
```

```python
# Method:1 Specifying scaling factors fx and fy
resized_cr=cv2.resize(cr,None,fx=0.1,fy=0.1)
plt.subplot(1,2,1);plt.imshow(resized_cr,cmap='gray')
```

```python
# Method:2 Specifying exact size size of output image
desired_width=100
desired_height=200
dim=(desired_width,desired_height)
resized_cr1=cv2.resize(cr,dsize=dim,interpolation=cv2.INTER_AREA)
plt.subplot(1,2,2);plt.imshow(resized_cr1,cmap='gray')
```

```python
img_NZ_rgb_flipped_hrz=cv2.flip(img,1)
img_NZ_rgb_flipped_vert=cv2.flip(img,0)
img_NZ_rgb_flipped_both=cv2.flip(img,-1)

plt.figure(figsize=(18,5))
plt.subplot(141);plt.imshow(img_NZ_rgb_flipped_hrz,cmap='gray');plt.title("Horizontal Flip")
plt.subplot(142);plt.imshow(img_NZ_rgb_flipped_vert,cmap='gray');plt.title("Vertical Flip")
plt.subplot(143);plt.imshow(img_NZ_rgb_flipped_both,cmap='gray');plt.title("Both Flip")
plt.subplot(144);plt.imshow(img,cmap='gray');plt.title("Original Flip")
```

```python
# image Augmentation=== Draw Line
img=cv2.imread("/content/drive/MyDrive/CV_Datasets/foster-lake.jpg",1)
imageLine=img.copy()
cv2.line(imageLine,(20,1000),(4000,10),(0,0,255),thickness=2,lineType=cv2.LINE_AA)
plt.imshow(imageLine[:,:,::-1])#bgr to rgb
```

```python
# image Augmentation=== Draw circle
img=cv2.imread("/content/drive/MyDrive/CV_Datasets/foster-lake.jpg",1)
imagecircle=img.copy()
cv2.circle(imagecircle,(3000,2000),100,(0,0,255),thickness=2,lineType=cv2.LINE_AA)
plt.imshow(imagecircle[:,:,::-1])#bgr to rgb
```

```python
# image Augmentation=== Draw Line
img=cv2.imread("/content/drive/MyDrive/CV_Datasets/foster-lake.jpg",1)
imageRectangle=img.copy()
cv2.rectangle(imageRectangle,(20,1000),(4000,10),(0,0,255),thickness=2,lineType=cv2.LINE_AA)
# ###########################left top , right bottom
plt.imshow(imageRectangle[:,:,::-1])#bgr to rgb
```

```python
imageText=img.copy()
text="hey"
fontscale=12
fontface=cv2.FONT_ITALIC
fontcolor=(0,0,0)
fontthickness=20
cv2.putText(imageText,text,(4000,1000),fontface,fontscale,fontcolor,fontthickness,cv2.LINE_AA)
plt.imshow(imageText[:,:,::-1])#bgr to rgb
```
