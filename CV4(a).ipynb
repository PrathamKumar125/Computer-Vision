# **4. a) write a program to implement Image Filtering (Edge Detection)**

```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img_bgr=cv.imread("/content/drive/MyDrive/CV_Datasets/OIP.jpg",cv.IMREAD_COLOR)
img_rgb=cv.cvtColor(img_bgr,cv.COLOR_BGR2RGB)

plt.imshow(img_rgb)
```

```python
# addition or brightness
import numpy as np
matrix=np.ones(img_rgb.shape,dtype="uint8")
img_rgb_brighter=cv.add(img_rgb,matrix)
img_rgb_darker=cv.subtract(img_rgb,matrix)

plt.figure(figsize=[18,5])
plt.subplot(131);plt.imshow(img_rgb_darker);plt.title("Darker")
plt.subplot(132);plt.imshow(img_rgb);plt.title("Original")
plt.subplot(133);plt.imshow(img_rgb_brighter);plt.title("Brighter")
```

```python
# Multiplication or Contrast
matrix1 = np.ones(img_rgb.shape)*0.8
matrix2 = np.ones(img_rgb.shape)*1.2
img_rgb_brighter = np.uint8(cv.multiply(np.float64(img_rgb), matrix1))
img_rgb_darker = np.uint8(cv.multiply(np.float64(img_rgb), matrix2))
plt.figure(figsize=[18, 5])
plt.subplot(131); plt.imshow(img_rgb_darker); plt.title("Higher contrast");
plt.subplot(132); plt.imshow(img_rgb);        plt.title("Original");
plt.subplot(133); plt.imshow(img_rgb_brighter); plt.title("Lower contrast");
```

```python
matrix1 = np.ones(img_rgb.shape)*0.8
matrix2 = np.ones(img_rgb.shape)*1.2
img_rgb_brighter = np.uint8(np.clip(cv2.multiply(np.float64(img_rgb), matrix2), 0, 255))
img_rgb_darker = np.uint8(cv2.multiply(np.float64(img_rgb), matrix1))
plt.figure(figsize=[18, 5])
plt.subplot(131); plt.imshow(img_rgb_darker); plt.title("Lower contrast");
plt.subplot(132); plt.imshow(img_rgb);        plt.title("Original");
plt.subplot(133); plt.imshow(img_rgb_brighter); plt.title("Higher contrast");
```

```python
# Image Thresholding
import cv2
img_read=cv.imread("/content/drive/MyDrive/CV_Datasets/OIP.jpg",cv.IMREAD_GRAYSCALE)
retval,img_thresh=cv2.threshold(img_read,50,255,cv2.THRESH_BINARY)
plt.figure(figsize=[18,5])

plt.subplot(121); plt.imshow(img_read,cmap="gray"); plt.title("Original");
plt.subplot(122); plt.imshow(img_thresh,cmap="gray"); plt.title("Thresholded");

print(img_thresh.shape)
```

```python
# Binary operations
img_rec=cv.imread("/content/drive/MyDrive/CV_Datasets/RECT.jpg",cv.IMREAD_GRAYSCALE)
img_rec=cv2.resize(img_rec,(400,150))

img_cir=cv.imread("/content/drive/MyDrive/CV_Datasets/CIRC.jpg",cv.IMREAD_GRAYSCALE)
img_cir=cv2.resize(img_cir,(400,150))

plt.subplot(121); plt.imshow(img_rec,cmap="gray"); plt.title("RECTANGLE");
plt.subplot(122); plt.imshow(img_cir,cmap="gray"); plt.title("CIRCLE");
print(img_rec.shape)
```

```python
result=cv.bitwise_and(img_rec,img_cir,mask=None)
plt.imshow(result,cmap="gray")
```

```python
result=cv.bitwise_or(img_rec,img_cir,mask=None)
plt.imshow(result,cmap="gray")
```

```python
result=cv.bitwise_xor(img_rec,img_cir,mask=None)
plt.imshow(result,cmap="gray")
```

```python
result=cv.bitwise_not(img_rec,img_cir,mask=None)
plt.imshow(result,cmap="gray")
```
